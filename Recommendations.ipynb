```python
import numpy as np
import pandas as pd
from typing import List, Dict
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
song_lyrics = pd.read_csv(r'C:\Users\aditi\Desktop\NLPProjects\Song Recommendations\songdata.csv')
song_lyrics.size
```

```python
song_lyrics.head(50)
```

```python
#Since this dataset is huge, take a random sample, because otherwise literally everything would be ABBA 
song_lyrics = song_lyrics.sample(n=5000, replace = True).reset_index(drop=True)
```

```python
song_lyrics
```

```python
#the link column is not necessary so let's remove it
# songs_modified = {'Artist':song_lyrics['artist'], 'Song':song_lyrics['song'], 'Lyrics':song_lyrics['text']}
# songs_modified
# df = pd.DataFrame(data=songs_modified)
# df
song_lyrics = song_lyrics.drop('link', axis = 1)
```

```python
song_lyrics
```

```python
#Clean the lyrics by getting rid of the \n tags
song_lyrics.columns = ['Artist', 'Song', 'Lyrics'] 
song_lyrics['Lyrics'] = song_lyrics['Lyrics'].str.replace(r'\n', '')
song_lyrics
```

```python
tfidf = TfidfVectorizer(analyzer='word', stop_words='english')
lyrics_matrix = tfidf.fit_transform(song_lyrics['Lyrics'])
lyrics_matrix
```

```python
cosine_similarities = cosine_similarity(lyrics_matrix) 
```

```python
similar_songs = {}
```

```python
for i in range(len(cosine_similarities)):
    closest_indices = cosine_similarities[i].argsort()[:-50:-1]  
    similar_songs[song_lyrics['Song'].iloc[i]] = [(cosine_similarities[i][x], song_lyrics['Song'][x], song_lyrics['Artist'][x]) for x in closest_indices][1:]
```

```python
similar_songs
```

```python
class SongRecommender:
    def __init__(self, matrix):
        self.matrix_similar = matrix
        
    def recommend(self, recommendation):
        song = recommendation['song']
        number_songs = recommendation['number_songs']
        recom_song = self.matrix_similar[song][:number_songs]
        self._print_message(song=song, recom_song=recom_song)

    def _print_message(self, song, recom_song):
        rec_items = len(recom_song)
        print(f'The {rec_items} recommended songs for {song} are:')
        for i in range(rec_items):
            print(f"Number {i+1}:")
            print(f"{recom_song[i][1]} by {recom_song[i][2]} with a similarity score of {round(recom_song[i][0], 3)}") 
            print("--------------------")
        
    
```

```python
recommendations = SongRecommender(similar_songs)
```

```python
recommendation = {
    "song": song_lyrics['Song'].iloc[10],
    "number_songs": 4 
}
recommendation
```

```python
recommendations.recommend(recommendation)
```
